### Ranaming

可以通过 rename 修改 column 和 index

```python
print(data.rename(columns={'Name':'Animal'}))
print(data.rename(index={2:'my favorite'}))
```

```python
data.rename_axis('ID',axis='rows').rename_axis('Name & Score',axis='columns')
```

```python
data1 = pd.read_csv('TempFolder/cat1.csv')
data2 = pd.read_csv('TempFolder/cat2.csv')
data1, data2
```

concat 会把相同的column相连接，不同的不连接，注意，是直接接在下面，不会横着接在后面

```python
pd.concat([data1,data2])
```

```python
data2 = data2.rename(columns={'Sex':'orzz'})
pd.concat([data1,data2])
```

```python
data2 = data2.rename(columns={'Name':'d3ac'})
pd.concat([data1,data2])
```

```python
data1 = pd.read_csv('TempFolder/cat1.csv')
data2 = pd.read_csv('TempFolder/cat2.csv')
```

在拼接的时候，因为 columns 的名字可能是相同的，所以拼接上了之后就会导致索引的时候出现问题，所以 rsuffix 和 lsuffix 至少得有一个，这样才能拼接成功

```python
data1.join(data2,rsuffix='_2')
```
