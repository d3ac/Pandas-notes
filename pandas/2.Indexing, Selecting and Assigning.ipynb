# Indexing, Seleting and Assigning

```python
import pandas as pd
```

```python
data = pd.read_csv('TempFolder/train.csv')
pd.set_option('display.max_rows',10)
data
```

### Indexing

可以使用 "native python" 索引方式，接下来以列 "Name" 为例子，可以直接索引得到一个Series类型的变量

```python
print(type(data.Name))
data.Name
```

也可以通过方括号进行索引，方括号索引的好处就是遇见保留字符的时候也会生效

```python
data['Name']
```

```python
print(data['Name'][4])
print(data.Name[4])
```

使用pandas更高级的索引，for more advanced operations, `loc` and `iloc` are the ones you're supposed to be using.

使用 `iloc` 索引第“3”行

```python
print(data.iloc[3])
```

`iloc` 和 `loc` 都是先行再列，和 "native python" 不一样（先列再行）

```python
print(data.iloc[:,3],end='\n\n')
print("The data in row 3 column 3 is \"{}\"".format(data.iloc[3,3]))
print("The type of it is \"{}\"".format(type(data.iloc[3,3])))
```

再次调出方便查看

```python
data
```

其他的索引方式，注意：$l:r$ 是 $[l,r)$ 的

```python
print(data.iloc[1:3,5],end='\n\n')
print(data.iloc[[0,2,4],[1,-1]])
```

接下来看通过名字进行索引，注意，名字在后，行数在前

```python
print(data.loc[1,'Age'])
```

```python
print(data.loc[:,['Age','Name']])
```

```python
print(data.loc[2:4,'Survived':'Sex'])
```

loc 和 iloc使用 `:` 的时候也有区别，iloc $[l,r)$，loc $[l,r]$

```python
print(data.iloc[1:3,3])
print('\n\n')
print(data.loc[1:3,'Name'])
```

接下来尝试条件索引

```python
data.Sex == "male"
```

再通过这样的方式就可以成功的筛选出 $Sex$ 为 $male$ 的行

```python
data.loc[data.Sex=="male"]
```

还可以设置多个条件，以及使用各种各样的索引方式

```python
data.loc[((data.loc[:,'Sex'] == 'male') & (data.iloc[:,11] == 'S'))]
```

当一列中要选择多个目标的时候，使用 `isin` 要比挨个写等于符号更快

```python
data.loc[data.loc[:,'Embarked'].isin(['S','C'])]
```

`isnull` 和 `notnull` 可以快速的选择出是否是 `null` 的行或者列

```python
data.loc[data.Cabin.notnull()]
```

还可以新增一列，当不存在这一列的时候便会自动加在最后一列，当存在的时候便会修改

```python
data['TryToAddId'] = range(len(data))
data
```
